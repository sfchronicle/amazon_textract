```python
new_year = []
for i,v in enumerate(df_all['year']):
    totalcount = list(df_all['year']).count(v)
    count = list(df_all['year'])[:i].count(v)
    new_year.append(v + "_" + str(count + 1) if totalcount >1 else v) 

new_year
```

```python
for i in range(len(new_year)):
  if len(list(new_year[i])) <= 4:
    new_year[i] = ''
```

```python
df_all['new_year'] = new_year
df_all = df_all[['description','dispostion','new_year']]
```

```python
columns = ['new_year','dispostion']
df_all[columns] = df_all[columns].replace('',np.nan).ffill()
df_all
```

```python
df_all = df_all.groupby(columns)['description'].agg(''.join).reset_index().reindex(df_all.columns, axis=1)
```
